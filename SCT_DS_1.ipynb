```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df=pd.read_csv('/content/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_3401680.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python


# Count the number of countries in each income group
income_group_counts = df['IncomeGroup'].value_counts()

# Plotting the bar graph
plt.figure(figsize=(10, 6))
plt.bar(income_group_counts.index, income_group_counts.values, color='skyblue')
plt.title('Number of Countries by Income Group')
plt.xlabel('Income Group')
plt.ylabel('Number of Countries')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
region_counts = df['Region'].value_counts()

# Plotting the bar graph
plt.figure(figsize=(10, 6))
plt.bar(region_counts.index, region_counts.values, color='purple')
plt.title('Number of Countries by Region')
plt.xlabel('Region')
plt.ylabel('Number of Countries')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

file_path = 'API_SP.POP.TOTL_DS2_en_csv_v2_3401680.csv'
data = pd.read_csv(file_path, skiprows=4)

# Extract the population data for the year 2023 (or the most recent available year)
population_2023 = data['2023'].dropna()

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(population_2023, bins=20, color='green', edgecolor='black')
plt.title('Population Distribution Across Countries in 2023')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()





```
